```python
import pandas as pd
import numpy as np
```

```python
df= pd.DataFrame({'season': ['Spring','Summer','Fall','Winter', np.nan]})
df
```

```python
pd.get_dummies(df['season'])
#칼럼명만 넣도록
```

```python
pd.get_dummies(df)
#프레임명_컬럼명으로 넣도록
```

```python
pd.get_dummies(df['season'], dummy_na=True)
#NaN 컬럼 별도 생성하기
```
