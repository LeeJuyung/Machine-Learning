```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from sklearn import datasets
df=pd.read_csv("data/contents.csv")
print(df.shape)
df.head()
```

```python
df = df.drop(['일자'], axis=1)
df
```

```python
df[["ID","시청컨텐츠"]].values.tolist()
```

```python
dataset = []
for name in df.ID.unique():
    name_mov = list(df[df['ID']==name]['시청컨텐츠'])
    dataset.append(name_mov)
dataset
```

```python
#TransactionEncoder는 리스트의 리스트로 구성된 트랜잭션 데이터를 Numpy 배열로 인코딩하기 위함
te=TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#지지도 확인
frequent_itemsets= apriori(df,min_support = 0.2, use_colnames=True)
frequent_itemsets
```

```python
#지지도 내림차순 정렬
frequent_itemsets.sort_values('support', ascending=False)
```

```python
#신뢰도 확인
#설명 = shift + tap
association_rules(frequent_itemsets, metric ="confidence", min_threshold=1)
```

```python
#향상도 확인
association_rules(frequent_itemsets, metric="lift",min_threshold=1)
```

```python
df_assoc = association_rules(frequent_itemsets, metric="lift", min_threshold=0.3)
df_assoc.sort_values('lift', ascending =False) #lift 기준 내림차순 정렬
```

지지도가 가장 높은 조합은 스타워즈 -> 어벤져스이다.(지지도 상위 42%)
신뢰도는 75%, 향상도는 1.32 = 1이상으로 추천할 만 함
전체 기록 중에 스타워즈와 어벤져스를 모두 시청한 비율은 42%이며 스타워즈를 본 사람 중 75%는 이후 어벤져스를 시청하고,
이를 이용하여 마케팅을 하면 이 연관분석 결과를 모를 때보다 효과가 32% 증가한다는 것을 의미한다.

향상도를 기준으로 세가지의 조합이 나온 것 중 타이타닉, 겨울왕국, 미녀와 야수의 향상도도 1 이상으로 추천할 만한 조합이다. 위 조합을 적용하여 마케팅을 하면 시청효과가 3.5배 상승한다.
